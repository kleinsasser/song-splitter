```python
'''
This file splits the training data of musdb18hq to ~11.5 second snippets
of each mixture.txt and vocals.txt file and saves the snippets to .npy
files of the processed .wav files
'''
import numpy as np
from numpy import savetxt
import librosa
import os

train_dir = './musdb18hq/train/'
train_file_dirs = os.listdir(train_dir)
train_file_dirs.remove('.DS_Store')

target_sr = 8192
SL = target_sr * 12 - 1 # this sample length makes the stft 128 frames
```

```python
X = []
Y = []
count = 0
for d in train_file_dirs:
    print(count, '/ 100 Processing:', d)
    count += 1

    wav_files = os.listdir(train_dir + d)
    
    # where this takes time
    mix, _ = librosa.load(train_dir + d + '/mixture.wav', target_sr)
    vocal, _ = librosa.load(train_dir + d + '/vocals.wav', target_sr)
    
    # make sure audio file length is a multiple of the SL
    trail = np.zeros(SL - mix.shape[0] % SL) 
    mix = np.hstack((mix, trail))
    vocal = np.hstack((vocal, trail))

    # split em
    for i in range(int(mix.shape[0] / SL)):
        X.append(mix[SL * i : SL * i + SL])
        Y.append(vocal[SL * i : SL * i + SL])

    # if count == 2: break

print("converting...")
X = np.array(X)
Y = np.array(Y)

print("saving...")
np.save('unet_inputs_raw.npy', X)
np.save('unet_targets_raw.npy', Y)

print("done")
```
