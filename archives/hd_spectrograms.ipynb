```python
import numpy as np
import librosa
import librosa.display
import os
import matplotlib.pyplot as plt
from scipy.io.wavfile import write
from time import time
```

```python
fp = 'audio-files/rational_gaze.wav'

mix, sr = librosa.load(fp)
sample = mix[10000:10512]
```

```python
write('sample.wav', sr, sample)
```

```python
S = librosa.stft(sample)
D = librosa.amplitude_to_db(np.abs(S), ref=np.max)
librosa.display.specshow(D, y_axis='linear')
plt.colorbar(format='%+2.0f dB')
plt.title('Linear-frequency power spectrogram')
plt.show()
```

```python
print(D.shape)
```

```python
start = time()
mix_ = librosa.istft(S)
print(time() - start)
```

```python
write('rg_reconstructed.wav', sr, mix_)
```
